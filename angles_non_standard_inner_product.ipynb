```python
import numpy as np
```

```python
def find_angle(a, x, y):
    angle = np.arccos((x.T @ a @ y) / (np.sqrt(x.T @ a @ x) * np.sqrt(y.T @ a @ y)))
    return round(angle.reshape(1)[0], 2), np.floor(np.rad2deg(angle.reshape(1)[0]))
```

```python
x = np.array([[1], [1]])
y = np.array([[-1], [1]])
a = np.array([[2, -1], [-1, 4]])
angle_in_rad, angle_in_deg = find_angle(a=a, x=x, y=y)
print(f"{angle_in_rad} rad ({angle_in_deg} deg)")
```
