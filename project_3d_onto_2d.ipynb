```python
import numpy as np
```

```python
x = np.array([[6],
              [0],
              [0]])
b1 = np.array([[1],
                [1],
                [1]])
b2 = np.array([[0],
               [1],
               [2]])
b = np.concatenate((b1, b2), axis=1)
coordinate = np.linalg.inv(b.T @ b) @ b.T @ x
projection_of_x = coordinate[0,0] * b1 + coordinate[1,0] * b2
projection_matrix = b @ np.linalg.inv(b.T @ b) @ b.T
rank_of_matrix = np.linalg.matrix_rank(projection_matrix)
print(f"(coordinate) = {coordinate}")
print(f"the projection of x onto U is {projection_of_x}")
print(f"The projection matrix is {projection_matrix}")
print(f"The rank of the projection matrix is {rank_of_matrix}")
```

```python
x = np.array([[3],
              [2],
              [2]])
b1 = np.array([[1],
                [0],
                [0]])
b2 = np.array([[0],
               [1],
               [1]])
b = np.concatenate((b1, b2), axis=1)
coordinate = np.linalg.inv(b.T @ b) @ b.T @ x
projection_of_x = coordinate[0,0] * b1 + coordinate[1,0] * b2
projection_matrix = projection_of_x @ projection_of_x.T
square_of_projection = np.sum(projection_of_x**2)
rank_of_matrix = np.linalg.matrix_rank(projection_matrix)
print(f"(coordinate) = {coordinate}")
print(f"the projeciton of x onto U is {projection_of_x}")
print(f"The projection matrix is 1/{square_of_projection} {projection_matrix}")
print(f"The rank of the projection matrix is {rank_of_matrix}")
```
