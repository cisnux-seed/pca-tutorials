```python
import numpy as np
```

```python
def cov(dataset, mean):
    variances = dataset-mean
    cov = (variances[:,0] @ variances[:,1])/np.size(dataset, axis=0)
    return cov

def variance(dataset, mean):
    return np.sum((dataset-mean)**2)/np.size(dataset)
```

```python
dataset = np.array([[1,2],[5,4],[-2,-3],[4,-2],[2,3],[8,-9]])
mean = np.mean(dataset,axis=0)
```

```python
print("covariance(x,y) = {:.2f}".format(cov(dataset,mean)))
```

```python
# transpose vector before append them both to become a matrix
# dataset = np.array([[1,5],[2,4]]).T
dataset = np.array([[1,2],[5,4]])
mean = np.mean(dataset,axis=0)
print(f"mean x and y: {mean}")
print(f"variance x {variance(dataset[:,0],mean[0])}")
print(f"variance y {variance(dataset[:,1],mean[1])}")
print("cov(x,y) and cov(y,x) = {:.2f}".format(cov(dataset,mean)))
```
