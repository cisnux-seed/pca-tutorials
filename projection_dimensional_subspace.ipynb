```python
import numpy as np
```

```python
b = np.array([[1],
              [2],
              [2]])
b_squares = np.sum(b**2)
projection_matrix = b @ b.T
print(f"1/{b_squares} {projection_matrix}")
```

```python
x = np.array([[1],
              [1],
              [1]])
b = np.array([[3],
              [0],
              [4]])
a = np.array([
  [9,0,12],
  [0,0,0],
  [12,0,16]
])
b_squares = np.sum(b**2)
projection_matrix = a @ x
print(f"1/{b_squares} {projection_matrix}")
```

```python
x = np.array([[1],
              [1],
              [1]])
projection = (1 / 9) * np.array([[5],
                                 [10], 
                                 [10]])
x_to_projection = x-projection
print(np.sqrt(np.sum(x_to_projection**2)))
```
