```python
import numpy as np
```

```python
b = np.array([[1],
              [2],
              [2]])
b_squares = np.sum(b**2)
projection_matrix = b @ b.T
print(f"1/{b_squares} {projection_matrix}")
```

```python
x = np.array([[1],
              [1],
              [1]])
b = np.array([[3],
              [0],
              [4]])
a = np.array([
  [9,0,12],
  [0,0,0],
  [12,0,16]
])
b_squares = np.sum(b**2)
projection_matrix = a @ x
print(f"1/{b_squares} {projection_matrix}")
```

```python
x = np.array([[1],
              [1],
              [1]])
projected_vector = 1/9 * np.array([[5],
                             [10],
                             [10]])
x_off_by_projected = x-projected_vector
print(f"distance between x and projected matrix {np.sqrt(np.sum(x_off_by_projected**2))}")
```
