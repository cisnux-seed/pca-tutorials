```python
import numpy as np
```

```python
x = np.array([[1],
              [-1],
              [3]])
m = np.array([
  [2,1,0],
  [1,2,-1],
  [0,-1,2]])
inner_product = x.T @ m @ x
print(f"length of x = sqrt({inner_product.reshape(1)[0]})")
```

```python
x = np.array([[1/2],
              [-1],
              [-1/2]])
y = np.array([[0],
              [1],
              [0]])
x_to_y = x-y
inner_product = x_to_y.T @ m @ x_to_y
print(f"distance between x and y = {inner_product.reshape(1)[0]}")
```

```python
x = np.array([[-1],
              [1]])
m = np.array([
  [5,-1],
  [-1,5]]) * 1/2
inner_product = x.T @ m @ x
print(f"length of x = sqrt({inner_product.reshape(1)[0]})")
```

```python
x = np.array([[4],
              [2],
              [1]])
y = np.array([[0],
              [1],
              [0]])
m = np.array([
  [2,1,0],
  [1,2,-1],
  [0,-1,2]])
x_to_y = x-y
inner_product = x_to_y.T @ m @ x_to_y
print(f"length of x = {np.sqrt(inner_product.reshape(1)[0])}")
```
