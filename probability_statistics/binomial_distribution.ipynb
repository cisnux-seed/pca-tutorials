```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_binomial_distributions_pmf(n, p, x_title = ""):
    pmf = np.zeros(n + 1)
    x = np.arange(n + 1)
    for i in range(n + 1):
        coeffecient = math.factorial(n) / (math.factorial(i) * math.factorial(n - i))
        print(coeffecient)
        pmf[i] = coeffecient * (p**i) * ((1 - p) ** (n - i))
    _, ax = plt.subplots()
    ax.bar(x, pmf)
    ax.set_xlabel(x_title)
    ax.set_ylabel('probability distribution')
    plt.show()
    return pmf
```

```python
# unbiased coin to find number of heads in 5 coins tosses
print(f"pmf of coin = {find_binomial_distributions_pmf(n=10, p=0.5, x_title='number of heads')}")
```

```python
# unbiased dices to find number of times we obtain the number 1
pmf = find_binomial_distributions_pmf(n=10, p=1/6, x_title='number of times the dice got 1')
sum_of_pmf = round(np.sum(pmf))
print(f"pmf of dices = {pmf}")
print(f"sum of pmf = {sum_of_pmf}")
```

```python
# unbiased dices to find number of heads in 5 coins tosses
print(f"pmf of dices = {find_binomial_distributions_pmf(n=5, p=1/6, x_title='number of times the dice got 1')}")
```

```python
# biased coin to find number of heads in 5 coins tosses
print(f"pmf of coin = {find_binomial_distributions_pmf(n=5, p=0.3, x_title='number of heads')}")
```
