```python
import numpy as np
```

```python
def len_of_vector(vector):
    r,c = vector.shape
    sum_of_squares = 0
    for i in range(r):
      for j in range(c):
          sum_of_squares += vector[i,j]**2
    return sum_of_squares,np.sqrt(sum_of_squares)
```

```python
x = np.array([
    [1],
    [-1],
    [3]
])
sum_of_squares,length = len_of_vector(x)
print(f"length of vector = sqrt({sum_of_squares}) = {length}")
```

```python
x = np.array([[3],
              [4]])
y = np.array([[1],
              [-1]])
_,len_of_x = len_of_vector(x)
_,len_of_y = len_of_vector(y)
angle = np.arccos((x.T@y).reshape(1)[0]/(len_of_x*len_of_y))
print(f"angle={angle}")
```

```python
x=np.array([[3],
            [4]])
y=np.array([[1],
            [1]])
_,vector_length=len_of_vector(x-y)
print(f"length of vector {vector_length}")
```

```python
x=np.array([[1],
            [2],
            [3]])
y=np.array([[-1],
            [0],
            [8]])
x_to_y=x-y
_,len_of_x = len_of_vector(x)
_,len_of_x_to_y = len_of_vector(x_to_y)
angle = np.arccos((x.T@x_to_y).reshape(1)[0]/(len_of_x*len_of_x_to_y))
print(f"angle = {angle}")
```
